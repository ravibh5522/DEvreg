```python
import os
from openai import OpenAI
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
client = OpenAI()
```

```python
def output(prompt):
  response = client.chat.completions.create(
  model="ft:gpt-3.5-turbo-1106:devrev-forge::8vOg7kcC",
  messages=[
    {"role": "system",
     "content": "output can be only spam or possitive_feedback or negetive_feedback or bugReport or bugCritical"},
    {
      "role": "user",
      "content": prompt
    }
  ],
  temperature=1,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
  return response.choices[0].message.content
```
